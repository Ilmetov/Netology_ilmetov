```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import scipy as sp
%matplotlib inline
```

Экспоненциальное распределение

```python

exp_rv = sts.expon()

x = np.linspace(0,5,1000)
y = np.linspace(1,3,1000)
z = np.linspace(0,10, 1000)
pdf = exp_rv.pdf(x)
pdf2 = exp_rv.pdf(y)
pdf3 = exp_rv.pdf(z)
cdf = exp_rv.cdf(x)
cdf2 = exp_rv.cdf(y)
cdf3 = exp_rv.cdf(z)
plt.figure(figsize=(10,10))
plt.plot(pdf)
plt.plot(pdf2)
plt.plot(pdf3)
plt.plot(cdf)
plt.plot(cdf2)
plt.plot(cdf3)
plt.title('Exponential distribution')
plt.legend(['pdf 0:5',  'pdf 1:3', 'pdf 0:10', 'cdf 0:5',  'cdf 1:3', 'cdf 0:10'])
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

Распределение Коши

```python
cauchy_rv = sts.cauchy()

x = np.linspace(0,5,1000)
y = np.linspace(1,3,1000)
z = np.linspace(0,10, 1000)
pdf = cauchy_rv.pdf(x)
pdf2 = cauchy_rv.pdf(y)
pdf3 = cauchy_rv.pdf(z)
cdf = cauchy_rv.cdf(x)
cdf2 = cauchy_rv.cdf(y)
cdf3 = cauchy_rv.cdf(z)
plt.figure(figsize=(10,10))
plt.plot(pdf)
plt.plot(pdf2)
plt.plot(pdf3)
plt.plot(cdf)
plt.plot(cdf2)
plt.plot(cdf3)
plt.title('Cauchy distribution')
plt.legend(['pdf 0:5',  'pdf 1:3', 'pdf 0:10', 'cdf 0:5',  'cdf 1:3', 'cdf 0:10'])
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

Распределение Парето

```python
pareto_rv = sts.pareto(2, loc=0, scale=1)
pareto_rv2 = sts.pareto(4, loc=1, scale=2)
pareto_rv3 = sts.pareto(2, loc=1, scale=3)

x = np.linspace(0,9,10)
pdf = pareto_rv.pdf(x)
pdf2 = pareto_rv2.pdf(x)
pdf3 = pareto_rv3.pdf(x)
cdf = pareto_rv.cdf(x)
cdf2 = pareto_rv2.cdf(x)
cdf3 = pareto_rv3.cdf(x)
plt.figure(figsize=(10,10))
plt.plot(pdf)
plt.plot(pdf2)
plt.plot(pdf3)
plt.plot(cdf)
plt.plot(cdf2)
plt.plot(cdf3)
plt.title('Pareto distribution')
plt.legend(['pdf 2, 0, 1',  'pdf 4, 1, 2', 'pdf 2, 1, 3', 'cdf 2, 0, 1',  'cdf 4, 1, 2', 'cdf 2, 1, 3'])
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```
