Задание 1
Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце 'region' пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим 'undefined'.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {

    'Центр': ['москва', 'тула', 'ярославль'],

    'Северо-Запад': ['петербург', 'псков', 'мурманск'],

    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']

}

Результат классификации запишите в отдельный столбец region.



```python
geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
# показалось, что такой вариант более "правильный", чтобы обработать вероятность двух ключевых слов, 
def get_key(row):
    geo = []
    row_split = str(row['keyword']).split()
    for each in row_split:
        for key, value in geo_data.items():
            if each in value:
                geo.append(key)
    #так и не придумал, какой есть вариант элегантнее, если работать со словарём относительно значений, а не ключа
    if geo==[]:
         geo.append('undefined')
    return geo
```

```python
#keywords['region']=keywords.apply(get_key(geo_data, keywords["keyword"]), axis=1)
keywords['region']=keywords.apply(get_key, axis=1)
keywords[keywords['keyword'].str.contains('москва', na=False)].head(10)
```

Задание 2
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
    - оценка 2 и меньше - низкий рейтинг
    - оценка 4 и меньше - средний рейтинг
    - оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
def rating_class (row):
    if row['rating'] == 2.0:
        r = 'низкий рейтинг'
    elif row['rating'] <= 4.0:
        r = 'средний рейтинг'
    elif row['rating'] in [4.5, 5.0]:
        r = 'высокий рейтинг'
    else:
        r = 'no rating'
    return r
```

```python
ratings['class'] = ratings.apply(rating_class, axis = 1)
ratings.head()
```

Задание 3
Посчитайте среднее значение Lifetime киноманов (пользователи, которые поставили 100 и более рейтингов). Под Lifetime понимается разница между максимальным и минимальным значением timestamp для каждого пользователя. Ответ дайте в днях.

```python
r = ratings.groupby('userId').agg(['min','max','count'])['timestamp']
r_filter = r[r['count']>100]
#r_filter['lifetime'] = r_filter.apply(lambda row: row['max'] - row['min'], axis=1)
r_filter['lifetime'] = (r_filter['max'] - r_filter['min'])/60/60/25
avg_lt = r_filter['lifetime'].mean()
avg_lt
```


Задание 4
Есть мнение, что "раньше снимали настоящее кино, не то что сейчас". Ваша задача проверить это утверждение, используя файлы с рейтингами фильмов из материалов занятия. Т. е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.

При этом мы не будем затрагивать субьективные факторы выставления этих рейтингов, а пройдемся по следующему алгоритму:

1. В переменную years запишите список из всех годов с 1950 по 2010.

2. Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:
    - для каждой строки пройдите по всем годам списка years
    - если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
    - если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год

3. Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец 'year'

4. Посчитайте средний рейтинг всех фильмов для каждого значения столбца 'year' и отсортируйте результат по убыванию рейтинга

```python
years = list(range(1950,2011))
movies = pd.read_csv('movies.csv')
```

```python
def production_year(row):
    year = ''
    for each in years:
        if str(each) in row['title']:
            year = str(each)
    if year =='':
        return '1900'
    else:
        return year
```

```python
movies['year'] = movies.apply(production_year, axis=1)
movies.head()
```

```python
movies_rating = ratings.merge(movies, on = 'movieId', how = 'left')
rating_result = movies_rating.groupby('year').agg(['mean'])['rating'].reset_index()
rating_result.sort_values('mean')
```

```python
rating_result['year_int'] = rating_result['year'].astype(str).astype(int)
rating_result.dtypes
```

```python
xrange = list(range(1950,2011,10))
plt.plot((rating_result['year_int']), rating_result['mean'])
#plt.autoscale(tight=True)
plt.xlim([1950,2010])
plt.show()
```

По таблице не слишком заметно, но по графику хорошо видно, что "раньше снимали лучше", однако после "провальных" 80-90-х наблюдается зародившаяся тенденция к исправлению ситуации :)
