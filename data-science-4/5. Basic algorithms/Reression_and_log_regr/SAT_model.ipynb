```python
y=data.univ_GPA
X=data.drop('univ_GPA', axis=1)

# from sklearn.preprocessing import StandardScaler
# X_sc=pd.DataFrame(StandardScaler().fit_transform(X))
# y_sc=pd.DataFrame(StandardScaler().fit_transform(X))

from sklearn.linear_model import LinearRegression

linreg=LinearRegression(normalize=True).fit(X,y)
y_pred = linreg.predict(X)

from sklearn.metrics import r2_score
print('R2 score: {}'.format(r2_score(y, y_pred)))
print('coefs: {}'.format(linreg.coef_))
print('intercept: {}'.format(linreg.intercept_))
```
